```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import gensim

story = []
for filename in os.listdir('Data'):

    f = open(os.path.join('Data',filename), encoding='latin-1')
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        sent = sent.lower()
        # stopwords removal
        processed_sent = [i for i in simple_preprocess(sent) if i not in gensim.parsing.preprocessing.STOPWORDS]
        story.append(processed_sent)
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,vector_size=100
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
